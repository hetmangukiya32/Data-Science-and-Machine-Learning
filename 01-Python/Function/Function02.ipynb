```python
##Temprature conversion
def convert_temp(temp,unit):
  """temp convert between celcius and fehrenheit"""
  if unit=='C':
    return temp * 9/5 + 32
  elif unit=="F":
        return (temp-32)*5/9 ## Fahrenheit to celsius
  else:
      return None

print(convert_temp(25,'C'))
print(convert_temp(77,'F'))
```

```python
##password strengh checker
# Check whether a password is strong or not

def check_password(password):

    if len(password) < 8:
        return False

    if not any(i.isupper() for i in password):
        return False

    if not any(i.islower() for i in password):
        return False

    if not any(i.isdigit() for i in password):
        return False

    if not any(i in "@#$%&*!" for i in password):
        return False

    return True


# Test
print(check_password("WeakPwd"))
print(check_password("Str0ngPwd!"))
```

```python
#Calculate total cost of item in shoping cart
def calculate_total_cost(cart):
    total_cost=0
    for item in cart:
        total_cost+=item['price']* item['quantity']

    return total_cost


## Example cart data

cart=[
    {'name':'giva','price':0.5,'quantity':4},
    {'name':'graps','price':0.3,'quantity':6},
    {'name':'Orange','price':0.7,'quantity':3}

]

## calling the function
total_cost=calculate_total_cost(cart)
print(total_cost)
```

```python
#palindrome program
def is_palindrome(s):
    s=s.lower().replace(" ","")
    return s==s[::-1]

print(is_palindrome("A man a plan a canal Panama"))
print(is_palindrome("Hello"))
```
