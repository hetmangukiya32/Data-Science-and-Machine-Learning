##### Introduction to Functions
Definition:

A function is a block of code that performs a specific task.
Functions help in organizing code, reusing code, and improving readability.


```python
##syntax
def function_name(parameters):
  """Docstring"""
  #Function body
  return expression
```

```python
##creare odd even function
def even_odd(num):
  if num%2==0:
    print("the number is even")
  else:
    print("the number is odd")
```

```python
##Call this function
even_odd(24)
```

```python
##Function with multiple parameters

def add(a,b):
    return a+b

result=add(2,4)
print(result)
```

```python
##Default parameters

def greet(name):
    print(f"hello {name}welcome to")

greet("Het")
```

```python
# Positional Arguments
# Arguments are assigned to parameters based on their position in the function call.

def greet(name, age):
    print(f"Hello {name}, you are {age} years old.")
    print(f"Hello {age}, you are {name} years old.")#error

greet("Alice", 25)
```

```python
# Keyword Arguments

# Arguments are assigned using the parameter names.

greet(name="Alice", age=25)
greet(age=25, name="Alice")
```

```python
def create_user(name, age, city):
    pass

create_user("John", 30, "London")               # positional
create_user(name="John", age=30, city="London") # keyword
create_user("John", age=30, city="London")      # mixed
```

```python
# *args  -> Accepts any number of positional arguments and stores them in a tuple.
# **kwargs -> Accepts any number of keyword arguments and stores them in a dictionary.
# This function prints all positional and keyword arguments passed by the user.

def print_details(*args, **kwargs):
    for val in args:
        print(f"Positional argument: {val}")

    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_details(
    ##Tuple
    1, 2, 3, 4, "Krish",
    #Dictonary
    name="Krish",
    age="32",
    country="India"
)
```

```python
### Return statements
def multiply(a,b):
    return a*b

multiply(2,3)
```

```python
## Return multiple parameters
def multiply(a,b):
    return a*b,a

multiply(2,3)
```
