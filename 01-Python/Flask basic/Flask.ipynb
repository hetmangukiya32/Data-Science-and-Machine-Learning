# **Flask**
Flask is a lightweight Python web framework used to build web applications, websites, and REST APIs. It provides the basic tools needed to create web applications quickly and easily.

Flask(__name__) → Creates a Flask application.

@app.route("/") → Maps a URL to a Python function.

def home() → Function that runs when the URL is opened.

return → Sends a response to the browser.

app.run(debug=True) → Starts the Flask development server with debug mode enabled.

```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def home():
    return "Hello, Flask!"

if __name__ == "__main__":
    app.run(debug=True)

  #o/p = hello flask
```

```python
Multiple Routes
from flask import Flask

app = Flask(__name__)

@app.route("/")
def home():
    return "Welcome to Flask!"

@app.route("/about")
def about():
    return "This is About Page."

@app.route("/contact")
def contact():
    return "Contact Us"

if __name__ == "__main__":
    app.run(debug=True)

#   o/p:-
#     http://127.0.0.1:5000/
# http://127.0.0.1:5000/about
# http://127.0.0.1:5000/contact
```

```python
from flask import Flask

app = Flask(__name__)

@app.route("/user/<name>")
def user(name):
    return f"Hello, {name}!"

if __name__ == "__main__":
    app.run(debug=True)

# URL
# http://127.0.0.1:5000/user/Het
# Output
# Hello, Het!
```

```python
# Addition API
from flask import Flask

app = Flask(__name__)

@app.route("/add/<int:a>/<int:b>")
def add(a, b):
    return f"Sum = {a + b}"

if __name__ == "__main__":
    app.run(debug=True)
```

```python
# JSON Response
from flask import Flask, jsonify

app = Flask(__name__)

@app.route("/student")
def student():
    data = {
        "name": "Het",
        "course": "Data Science",
        "age": 21
    }
    return jsonify(data)

if __name__ == "__main__":
    app.run(debug=True)
```

```python
# Simple HTML Page
from flask import Flask

app = Flask(__name__)

@app.route("/")
def home():
    return """
    <h1>Welcome to Flask</h1>
    <p>This is my first Flask application.</p>
    """

if __name__ == "__main__":
    app.run(debug=True)
```

```python
# Receive User Input
from flask import Flask, request

app = Flask(__name__)

@app.route("/greet")
def greet():
    name = request.args.get("name")
    return f"Hello {name}"

if __name__ == "__main__":
    app.run(debug=True)
```
