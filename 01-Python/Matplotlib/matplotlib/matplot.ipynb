```python
#Movies data
# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]
oscar_years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

plt.plot(oscar_years, oscar_revenue)
plt.title("Oscar Years vs Revenue(in $M)")
plt.xlabel("Years")
plt.ylabel("Revenue(in $M)")
```

```python
# Plot multiple
# Non-Oscar Winner Movies
non_oscar_movies = ["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_years = [2008, 2009, 2010, 2011, 2012]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M

plt.plot(oscar_years, oscar_revenue, label="Oscar Winners")
plt.plot(non_oscar_years, non_oscar_revenue, label="Non-Oscar Winners")
plt.title("Years vs Revenue(in $M)")
plt.xlabel("Years")
plt.ylabel("Revenue(in $M)")
plt.legend()
# plt.legend(loc="upper right")
# plt.legend(loc="upper left")
# plt.legend(loc="center")
```

```python
# using Format Strings
plt.plot(oscar_years, oscar_revenue, "o-b", label="Oscar Winners")
plt.plot(non_oscar_years, non_oscar_revenue, ">--g", label="Non-Oscar Winners")

plt.plot(oscar_years, oscar_revenue, color="black", linestyle="-", marker="o",label="Oscar Winners")
plt.plot(non_oscar_years, non_oscar_revenue, color="red", linestyle="--", marker=">", linewidth=5, label="Non-Oscar Winners")
```

```python
# Other Methods & Attributes
print(plt.style.available)
plt.plot(oscar_years, oscar_revenue, label="Oscar Winners")
plt.style.use('dark_background')
plt.style.use('ggplot')

plt.grid(True)

with plt.xkcd():
    plt.plot(oscar_years, oscar_revenue, "o-b", label="Oscar Winners")
    plt.plot(non_oscar_years, non_oscar_revenue, ">--g", label="Non-Oscar Winners")
    plt.xlabel("Years")
    plt.ylabel("Revenue(in $M)")
    plt.legend()
```

```python
# Bar charts
movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]
years = [2008, 2009, 2010, 2011, 2012]
revenue = [1005, 170, 427, 133, 232]  # in $M

plt.bar(years, revenue, color="purple")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
plt.title("Yearly revenue of movies")

# Add labels on top of bars
for i in range(len(revenue)):  ##length of revenue=5
    plt.text(years[i], revenue[i]+20, str(revenue[i]), ha="center")
    # text(pos of text acc to x-axis, pos of y, text, alignment)

plt.ylim(0, max(revenue)+200) #lower, upper limit - add space
```

```python
# comparing data with bar charts
oscar_movies = ["The Dark Knight","The Hurt Locker","The King's Speech","The Artist","Argo"]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

non_oscar_movies = ["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M

years = [2008, 2009, 2010, 2011, 2012]

# Plotting Bar Chart
import numpy as np
x = np.arange(len(years))  #x = [0,1,2,3,4]
width = 0.4

plt.bar(x-width/2, oscar_revenue, width, label="Oscar")  #[-0.2,0.8,1.8,2.8,3.8]
plt.bar(x+width/2, non_oscar_revenue, width, label="Non-Oscar")

plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
plt.title("Yearly revenue of Oscar vs Non-Oscar movies")
plt.legend()
plt.tight_layout()

plt.xticks(x, years) # to show x-axis labels
```

```python
# Scatter Plots
people = ["Person A", "Person B", "Person C", "Person D", "Person E",
          "Person F", "Person G", "Person H", "Person I", "Person J"]
age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
blood_pressure = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]

# plot
# plt.scatter(age, blood_pressure, color="red")

plt.title("Age vs Blood Pressure")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.grid()

# Customizing plots
colors = ["green" if bp<135 else "red" for bp in blood_pressure]
sizes = blood_pressure
plt.scatter(age, blood_pressure, color=colors, s=sizes, marker='^', alpha=0.5)

# plt.scatter(age, blood_pressure, c=blood_pressure, cmap="plasma")
plt.scatter(age, blood_pressure, c=blood_pressure, cmap="OrRd")# makes more sense
# plt.scatter(age, blood_pressure, c=[5]*len(age), cmap="plasma")
plt.colorbar(label="BP")

# Annotation
for i in range(len(age)):
    plt.annotate(people[i], xy=(age[i], blood_pressure[i]))
plt.xlim(min(age), max(age)+10)
```
