```python
#Abstraction
# Hiding implementation details
from abc import ABC, abstractmethod

class Animal(ABC):

    @abstractmethod
    def sound(self):
        pass


class Dog(Animal):

    def sound(self):
        print("Woof!")


class Cat(Animal):

    def sound(self):
        print("Meow!")


dog = Dog()
dog.sound()

cat = Cat()
cat.sound()
```

```python
from abc import ABC,abstractmethod

class Vehicle(ABC):
    def drive(self):
        print("The vehicle is used for driving")

    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        print("Car enginer started")

def operate_vehicle(vehicle):
    vehicle.start_engine()
    vehicle.drive()

car=Car()
operate_vehicle(car)
```
