# List vs Tuple in Python

## List
- Ordered collection
- Mutable (can be changed)
- Uses square brackets `[]`

```python
# List Example

fruits = ["Apple", "Banana", "Mango"]

print("Original List:", fruits)

# Add item
fruits.append("Orange")

# Change item
fruits[1] = "Grapes"

# Remove item
fruits.remove("Apple")

print("Updated List:", fruits)
```

**Output**
```
Original List: ['Apple', 'Banana', 'Mango']
Updated List: ['Grapes', 'Mango', 'Orange']
```

---

## Tuple
- Ordered collection
- Immutable (cannot be changed)
- Uses parentheses `()`

```python
# Tuple Example

fruits = ("Apple", "Banana", "Mango")

print("Tuple:", fruits)

# Access element
print("First Item:", fruits[0])

# This will give an error
# fruits[1] = "Grapes"
```

**Output**
```
Tuple: ('Apple', 'Banana', 'Mango')
First Item: Apple
```

---

## Example

```python
# List
numbers = [10, 20, 30]
numbers.append(40)
print(numbers)

# Tuple
numbers = (10, 20, 30)
print(numbers)
```

**Output**
```
[10, 20, 30, 40]
(10, 20, 30)
```

## When to Use?

- **Use List** → When data needs to change.
- **Use Tuple** → When data should remain fixed.

# ***Set***
mutable
does not Allow duplicate values


````python
# Create a Set
#set is unordered
numbers = {10, 20, 30, 40}

print(numbers)
````

````python
# Add an Element
numbers = {10, 20, 30}

numbers.add(40)

print(numbers)
````

````python
#Add Multiple Elements
numbers = {10, 20}

numbers.update([30, 40, 50])

print(numbers)
````

````python
#Remove an Element
numbers = {10, 20, 30}

numbers.remove(20)

print(numbers)
````

````python
# Remove an Element Without Error
numbers = {10, 20, 30}

numbers.discard(100)

print(numbers)
````

````python
# Remove a Random Element
numbers = {10, 20, 30}

removed = numbers.pop()

print("Removed:", removed)
print(numbers)
````

````python
# Clear the Set
# numbers = {10, 20, 30}

numbers.clear()

print(numbers)
````

````python
# Copy a Set
numbers = {10, 20, 30}

new_set = numbers.copy()

print(new_set)
````

````python
# Union of Two Sets
a = {1, 2, 3}
b = {3, 4, 5}

print(a.union(b))
````

````python
# Intersection of Two Sets
a = {1, 2, 3}
b = {2, 3, 4}

print(a.intersection(b))
````

````python
# Difference of Two Sets
a = {1, 2, 3}
b = {2, 3, 4}

print(a.difference(b))
````

````python
# Symmetric Difference
a = {1, 2, 3}
b = {2, 3, 4}

print(a.symmetric_difference(b))
````

````python
# Check Subset
a = {1, 2}
b = {1, 2, 3, 4}

print(a.issubset(b))
````

````python
# Check Superset
a = {1, 2, 3, 4}
b = {1, 2}

print(a.issuperset(b))
````

````python
# Check Disjoint Sets
a = {1, 2}
b = {3, 4}

print(a.isdisjoint(b))
````

````python
# Remove Duplicate Values Using Set
numbers = [10, 20, 20, 30, 30, 40]

unique = set(numbers)

print(unique)
````

````python
# Find Common Elements
a = {10, 20, 30, 40}
b = {30, 40, 50, 60}

common = a.intersection(b)

print(common)
````

````python
# Find Unique Elements
a = {10, 20, 30}
b = {20, 30, 40}

unique = a.symmetric_difference(b)

print(unique)
````

````python
# Check if Element Exists
numbers = {10, 20, 30}

if 20 in numbers:
    print("Found")
else:
    print("Not Found")
````

````python
# Iterate Through a Set
numbers = {10, 20, 30}

for num in numbers:
    print(num)
````

Dictionary in Python

A Dictionary stores data in Key : Value pairs.

Ordered
Mutable
Keys must be unique
Values can be duplicate
Uses curly braces {}

````python
student = {
    "name": "Het",
    "age": 20,
    "city": "Ahmedabad"
}

print(student)
````

````python
# Access Value
student = {
    "name": "Het",
    "age": 20
}

print(student["name"])
````

````python
# get()
# Returns the value of a key. If the key doesn't exist, it returns None (or a default value).

student = {
    "name": "Het",
    "age": 20
}

print(student.get("age"))
````

````python
# Add a New Key
student = {
    "name": "Het"
}

student["city"] = "Ahmedabad"

print(student)
````

````python
# Update Value
student = {
    "name": "Het",
    "age": 20
}

student["age"] = 21

print(student)
````

````python
# update()
# Updates one or more key-value pairs.

student = {
    "name": "Het"
}

student.update({
    "age": 20,
    "city": "Ahmedabad"
})

print(student)
````

````python
# keys()
# Returns all keys.

student = {
    "name": "Het",
    "age": 20
}

print(student.keys())
````

````python
# values()
# Returns all values.

student = {
    "name": "Het",
    "age": 20
}

print(student.values())
````

````python
# items()
# Returns all key-value pairs.

student = {
    "name": "Het",
    "age": 20
}

print(student.items())
````

````python
# pop()

# Removes a specific key.

student = {
    "name": "Het",
    "age": 20
}

student.pop("age")

print(student)
````

````python
# copy()

# Creates a copy of the dictionary.

student = {
    "name": "Het",
    "age": 20
}

new_student = student.copy()

print(new_student)
````

````python
# Check Key Exists
student = {
    "name": "Het",
    "age": 20
}

if "name" in student:
    print("Key Found")
else:
    print("Key Not Found")
````
