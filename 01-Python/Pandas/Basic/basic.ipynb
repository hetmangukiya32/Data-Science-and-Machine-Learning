**Pandas**
What is Pandas?

Pandas is a Python library used to work with table data.

Pandas helps us to:

Read CSV and Excel files
Store data in tables
Filter data
Sort data
Find averages
Handle missing values
Analyze large datasets

Most Data Science and Machine Learning projects use Pandas.



```python
# Q1 What is Pandas?
# Pandas is a Python library used for data manipulation and data analysis.

# Q2. Why do we use Pandas?
# To read, clean, filter, sort, analyze, and manipulate tabular data efficiently.

# Q3. Why do we write import pandas as pd?
# pd is a commonly used alias that makes the code shorter and easier to write.
```

```python
# Series (1D Data)
import pandas as pd
series_name = pd.Series([5,6,8,9,64])

print(series_name)
```

```python
# Access Values
import pandas as pd

marks = pd.Series([85, 90, 78, 95])

print(marks[0])
print(marks[2])
```

```python
# Negative Index
import pandas as pd

marks = pd.Series([85, 90, 78, 95])

print(marks.iloc[-1])
```

```python
import pandas as pd

marks = pd.Series([85, 90, 78])

print(type(marks))
```

```python
# Length of Series
import pandas as pd

marks = pd.Series([85, 90, 78, 95])

print(len(marks))
```

```python
# Find Maximum
import pandas as pd

marks = pd.Series([85, 90, 78, 95])

print(marks.max())
```

```python
# Find Minimum
import pandas as pd

marks = pd.Series([85, 90, 78, 95])

print(marks.min())
```

```python
# Find Sum
import pandas as pd

marks = pd.Series([85, 90, 78, 95])

print(marks.sum())
```

```python
# Find Average
import pandas as pd

marks = pd.Series([85, 90, 78, 95])

print(marks.mean())
```

```python
# 1. Index
# Every value in a Series has an index.
import pandas as pd

marks = pd.Series([85, 90, 78, 95])

print(marks.index)
```

```python
# 2. Custom Index
# Instead of numbers, we can create our own indexes.
import pandas as pd

marks = pd.Series(
    [85,90,78],
    index=["Math","Science","English"]
)

print(marks)
```

```python
# Access Using Custom Index
import pandas as pd

marks = pd.Series(
    [85,90,78],
    index=["Math","Science","English"]
)

print(marks["Science"])
```

```python
# 3. Series Name

# A Series can have a name.

import pandas as pd

marks = pd.Series(
    [85,90,78],
    name="Student Marks"
)

print(marks)
print(marks.name ) ##only name
```

```python
# 4. Values Attribute
# Returns all values.
import pandas as pd

marks = pd.Series([85,90,78])

print(marks.values)
##returns numpy array
```

```python
# Check Type
import pandas as pd
marks = pd.Series([85,90,78])

print(type(marks.values))
```

```python
# 5. Data Type (dtype)
import pandas as pd

marks = pd.Series([85,90,78])

print(marks.dtype)
```

```python
# String Example
import pandas as pd

cities = pd.Series(["Surat","Rajkot"])

print(cities.dtype)
```

```python
# Float Example
import pandas as pd
salary = pd.Series([1000.5,2000.5])

print(salary.dtype)
```

```python
# Shape

# Shape tells us number of rows.
import pandas as pd

marks = pd.Series([85,90,78,95])

print(marks.shape)
```

```python
# Size
# Example
import pandas as pd

marks = pd.Series([85,90,78,95])

print(marks.size)
```

```python
# Number of Dimensions
# Example
import pandas as pd

marks = pd.Series([85,90,78])

print(marks.ndim)
```

```python
# Head()
# Shows first rows.
import pandas as pd

numbers = pd.Series([10,20,30,40,50,60])

print(numbers.head())
print(numbers.head(3)) ##First 3 rows
```

```python
# Tail()
# Shows last rows.

import pandas as pd

numbers = pd.Series([10,20,30,40,50,60])

print(numbers.tail())
print(numbers.tail(2))
```

```python
# Real Example
import pandas as pd

students = pd.Series(
    [85,90,78,95],
    index=["Het","Rahul","Amit","Jay"],
    name="Marks"
)

print("Series Name:", students.name)
print("Indexes:", students.index)
print("Values:", students.values)
print("Shape:", students.shape)
print("Size:", students.size)
print("Dimensions:", students.ndim)
```

DataFrame (2D **Data**)

⭐ DataFrame is the heart of Pandas.
Almost every Data Science, Machine Learning, and Data Analysis project uses DataFrames.

```python
# A DataFrame is a 2-dimensional table made up of rows and columns.
# Create DataFrame Using Dictionary
import pandas as pd

student = {
    "Name": ["Het", "Rahul", "Amit"],
    "Age": [21, 22, 20],
    "City": ["Surat", "Ahmedabad", "Rajkot"]
}

df = pd.DataFrame(student)

print(df)
```

```python
# DataFrame Using List of Lists
import pandas as pd

data = [
    ["Het",21,"Surat"],
    ["Rahul",22,"Ahmedabad"],
    ["Amit",20,"Rajkot"]
]

df = pd.DataFrame(data,
                  columns=["Name","Age","City"])

print(df)
```

```python
# DataFrame From List of Dictionaries
import pandas as pd

students = [
    {"Name":"Het","Age":21,"City":"Surat"},
    {"Name":"Rahul","Age":22,"City":"Ahmedabad"},
    {"Name":"Amit","Age":20,"City":"Rajkot"}
]

df = pd.DataFrame(students)

print(df)
```

```python
# DataFrame From NumPy Array
import pandas as pd
import numpy as np

arr = np.array([
    [10,20],
    [30,40],
    [50,60]
])

df = pd.DataFrame(arr,
                  columns=["Math","Science"])

print(df)
```

```python
import pandas as pd

student = {
    "Name":["Het","Rahul"]
}

df = pd.DataFrame(student)

print(type(df))
print(df.shape)
print(len(df))
print(len(df.columns))
print(df.ndim)
print(df.size)
print(df.index)
```
