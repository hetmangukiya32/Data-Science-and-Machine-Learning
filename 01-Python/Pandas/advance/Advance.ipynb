```python
import pandas as pd

student = {
    "Name": ["Het", "Rahul", "Amit"],
    "Age": [21, 22, 20],
    "City": ["Surat", "Ahmedabad", "Rajkot"]
}

df = pd.DataFrame(student)

print(df)
# loc = Label → (row label, column name)
# iloc = Integer Location → (row position, column position)

# Row Label (Index) = 0, 1, 2 (or custom labels like S1, S2, S3)
# Column Name = Name, Age, City
```

```python
df.loc[0:3]
```

```python
# What does iloc use?
# iloc doesn't use names.
# It uses positions.
print(df.iloc[1, 0])
```

```python
import pandas as pd

student = {
    "Name": ["Het", "Rahul", "Amit"],
    "Age": [21, 22, 20],
    "City": ["Surat", "Ahmedabad", "Rajkot"],
    "Marks": [90, 80, 85]
}

df = pd.DataFrame(student)

print(df)
print(df["Age"])
print(df["Marks"])
print(df[["Name", "Age"]])
print(type(df["Name"]))
print(df["Age"] > 20)
print(df["City"] == "Surat")      #Creates True/False values
# Now Filter
print(df[df["City"] == "Surat"])   #return only matching rows
```

```python
# What is nunique()?
# nunique() means:
# Number of Unique Values
# It counts how many different (unique) values are present in a column.
import pandas as pd

data = {
    "City": ["Surat", "Ahmedabad", "Surat", "Rajkot", "Rajkot"]
}

df = pd.DataFrame(data)

# print(df)
print(df["City"].nunique())
print(df["City"].unique())
```
