# **Memory management**

```python
#Refernce count

import sys

# Create a list
numbers = [1, 2, 3]

# Print reference count
print(sys.getrefcount(numbers))

# Create another reference
copy_numbers = numbers

# Reference count increases
print(sys.getrefcount(numbers))

# Delete one reference
del copy_numbers

# Reference count decreases
print(sys.getrefcount(numbers))
```

```python
# Memory Allocation
# Python allocates memory automatically

a = 100
name = "Python"
nums = [1, 2, 3]

print(a)
print(name)
print(nums)
```

```python
# Garbage Collection
import gc

# Create an object
data = [10, 20, 30]

# Delete the object
del data

# Free unused memory
gc.collect()

print("Garbage Collection Done")
```

```python
# Enable and Disable GC
import gc

# Check GC status
print(gc.isenabled())

# Disable GC
gc.disable()

print(gc.isenabled())

# Enable GC
gc.enable()

print(gc.isenabled())
```

```python
# Function Memory
# Local variables are removed after function ends

def demo():
    nums = [1, 2, 3]
    print(nums)

demo()

print("Function Finished")
```

```python
# Circular Reference
import gc

class Person:
    pass

# Create two objects
p1 = Person()
p2 = Person()

# Create circular reference
p1.friend = p2
p2.friend = p1

# Delete references
del p1
del p2

# Remove unused objects
gc.collect()

print("Memory Cleaned")
```

```python
# Memory Size
import sys

num = 100
text = "Python"
lst = [1, 2, 3]

# Print object size
print(sys.getsizeof(num))
print(sys.getsizeof(text))
print(sys.getsizeof(lst))
```

```python
# Object Identity
# Both variables point to the same object

a = [1, 2, 3]
b = a

print(id(a))
print(id(b))

print(a is b)
```

```python
# Delete Variable
name = "Het"

print(name)

# Delete variable
del name

# print(name)  # NameError
```

```python
# Complete Demo
import sys
import gc

# Create object
data = [10, 20, 30]

# Memory address
print(id(data))

# Object size
print(sys.getsizeof(data))

# Reference count
print(sys.getrefcount(data))

# Create another reference
copy = data

print(sys.getrefcount(data))

# Delete one reference
del copy

print(sys.getrefcount(data))

# Delete object
del data

# Run garbage collector
gc.collect()

print("Program Finished")
```
