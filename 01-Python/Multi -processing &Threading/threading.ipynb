Multithreading
Definition

Multithreading means creating multiple threads inside a single process.

A thread is the smallest unit of execution.

Diagram
One Process
│
├── Thread 1
├── Thread 2
├── Thread 3
└── Thread 4

All threads share the same memory and resources.

Example

A web browser uses multiple threads:

UI Thread
Download Thread
Audio Thread
Video Thread

All these threads work together inside one process

Advantages
Fast to create
Uses less memory
Threads share memory
Best for I/O-bound tasks

Disadvantages
Shared memory can cause race conditions
In Python, the GIL (Global Interpreter Lock) limits true parallel execution for CPU-bound tasks.


2. Multiprocessing
Definition

Multiprocessing means creating multiple processes.

Each process has its own memory and resources.

Diagram
Process 1
│
└── Thread

Process 2
│
└── Thread

Process 3
│
└── Thread

Advantages
Uses multiple CPU cores
Best for CPU-intensive tasks
One process crashing usually does not affect the others

Disadvantages
Uses more memory
Slower to create than threads
Communication between processes is more complex

```python
import threading
import time

def task(name):
    for i in range(3):
        print(f"{name} is running")
        time.sleep(1)

t1 = threading.Thread(target=task, args=("Thread-1",))
t2 = threading.Thread(target=task, args=("Thread-2",))
# Creates the  thread.

# target=task
#       The thread will execute the task() function.
# args=("Thread-1",)
#       Passes "Thread-1" as the name argument.
#       The comma is required because args must be a tuple.

t1.start()
t2.start()

t1.join()
# Main program waits until Thread-1 finishes.
t2.join()

print("Finished")
```

```python
from multiprocessing import Process
import time

def task(name):
    for i in range(3):
        print(f"{name} is running")
        time.sleep(1)

p1 = Process(target=task, args=("Process-1",))
p2 = Process(target=task, args=("Process-2",))

p1.start()
p2.start()

p1.join()
p2.join()

print("Finished")

# Main Program
#       │
#       ▼
# Create Process-1
# Create Process-2
#       │
#       ▼
# Start Both Processes
#       │
#       ▼
# Run task()
#       │
#       ▼
# join()
#       │
#       ▼
# Finished
```

```python
## PRocesses that run in parallel
### CPU-Bound Tasks-Tasks that are heavy on CPU usage (e.g., mathematical computations, data processing).
## PArallel execution- Multiple cores of the CPU

import multiprocessing

import time

def square_numbers():
    for i in range(5):
        time.sleep(1)
        print(f"Square: {i*i}")

def cube_numbers():
    for i in range(5):
        time.sleep(1.5)
        print(f"Cube: {i * i * i}")

if __name__=="__main__":
#   Why do we use this?
# This block ensures that the multiprocessing code runs only when this file is executed

    ## create 2 processes
    p1=multiprocessing.Process(target=square_numbers)
    p2=multiprocessing.Process(target=cube_numbers)
    t=time.time()

    ## start the process
    p1.start()
    p2.start()

    ## Wait for the process to complete
    p1.join()
    p2.join()

    finished_time=time.time()-t
    print(finished_time)
```

```python
import multiprocessing
import math
import time

def factorial(num):
    print(f"Factorial of {num} = {math.factorial(num)}")

if __name__ == "__main__":

    numbers = [50000, 60000, 70000]

    processes = []

    start = time.time()

    # Create processes
    for num in numbers:
        p = multiprocessing.Process(target=factorial, args=(num,))
        processes.append(p)

    # Start all processes
    for p in processes:
        p.start()

    # Wait for all processes to finish
    for p in processes:
        p.join()

    end = time.time()

    print(f"Execution Time: {end - start:.2f} seconds")
```
