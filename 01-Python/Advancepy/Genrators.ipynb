```python
##Genrators
##it is a special function that give one value at at a time
def numbers():
    yield 10
    yield 20
    yield 30

gen = numbers()

print(next(gen))
print(next(gen))
print(next(gen))
```

```python
def demo():
    print("Start")
    yield 1

    print("Middle")
    yield 2

    print("End")
    yield 3

g = demo()

print(next(g))
print(next(g))
print(next(g))
```

```python
##return and yield
# How it works
# The function starts.
# It creates the entire list [10, 20, 30].
# It returns the list.
# The function stops immediately.

Once return is executed, the function ends.
def numbers():
    return [10, 20, 30]

result = numbers()
print(result)
```

```python
# How it works
# yield 10 → Returns 10 and pauses the function.
# next(gen) again → Continues from where it paused and returns 20.
# next(gen) again → Returns 30.

# The function does not stop after yield. It pauses and continues later.
def numbers():
    yield 10
    yield 20
    yield 30

gen = numbers()

print(next(gen))
print(next(gen))
print(next(gen))
```
