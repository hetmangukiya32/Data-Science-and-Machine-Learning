```python
##iterators
text = "Python"

it = iter(text)

print(next(it))
print(next(it))
print(next(it))
```

```python
my_list=[1,2,3,4,5,6]
for i in my_list:
    print(i)
```

```python
## Iterator
iterator=iter(my_list)
print(type(iterator))
```

```python
try:
    print(next(iterator))
except StopIteration:
    print("There are no elements in the iterator")
```
