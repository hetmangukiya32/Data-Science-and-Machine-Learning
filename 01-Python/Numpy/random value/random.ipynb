# NumPy Random Module
The NumPy Random Module is used to generate random numbers.

It is commonly used in:

Machine Learning
Data Science
Simulations
Testing

```python
import numpy as np
print(np.random.rand())
```

```python
#Multiple values
print(np.random.rand(2,3))
```

```python
# randn()

# Generates random numbers from the standard normal distribution.
# Mean = 0
# Standard Deviation = 1
print(np.random.randn(3))
```

```python
##genrate random integers
print(np.random.randint(1,10))
print(np.random.randint(1,10,size=(2,3)))
```

```python
# random()
# Returns random float numbers between 0 and 1.
print(np.random.random())
```

```python
# choice()
# Selects a random value from a list.

arr = ["Apple", "Banana", "Mango"]

print(np.random.choice(arr))
```

```python
# seed()
# Produces the same random numbers every time.

np.random.seed(10)

print(np.random.randint(1,100))
```

```python
# shuffle()
# Shuffles the original array.

arr = np.array([1,2,3,4,5])
np.random.shuffle(arr)

print(arr)
```

```python
# permutation()
# Returns a shuffled copy of the array.

arr = np.array([1,2,3,4,5])

x = np.random.permutation(arr)

print(arr)
print(x)
```

```python
# Sort
# np.sort() sorts the array in ascending order.

arr = np.array([5, 2, 8, 1])

print(np.sort(arr))
```

```python
arr = np.array([[3,1],[4,2]])

print(np.sort(arr))
##By default row wise sorting
```

```python
# Search
# np.where() finds the index of matching values.

arr = np.array([10,20,30,20,40])

print(np.where(arr == 20))
```

```python
#search even numnbers

arr = np.array([1,2,3,4,5,6])
print(np.where(arr % 2 == 0))
```

```python
# Filter
# Filtering means selecting values based on a condition.

arr = np.array([10,20,30,40,50])

x = arr[arr > 25]

print(x)
```

```python
# Filter Even Numbers
arr = np.array([1,2,3,4,5,6])
x = arr[arr % 2 == 0]

print(x)
```
